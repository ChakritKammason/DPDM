## 2. mean  collumn 'city_name'  Missing 

```
all_mean = new_table.dropna(subset = ['avg_price'])['avg_price'].mean()
all_mean
```

## 3.  list 

```
all_cities_list = []
for each_city in sorted(list(all_city)):
  temp_table = new_table[new_table['city_name'] == each_city]
  temp_table[['avg_price']] = temp_table[['avg_price']].fillna(new_table[new_table['city_name'] == each_city].dropna()['avg_price'].mean())
  all_cities_list.append(temp_table)  
```

```
pd.concat(all_cities_list).fillna(all_mean)
```

```
no_missing = pd.concat(all_cities_list).fillna(all_mean)
no_missing.head()
```

# Smooth data (binning)

## HW10  Binning smooth by mean  'avg_price'  no_missing  50  (bins = 50)

panda

 30  

```
import math
each_bin = no_missing.shape[0]/50
print(math.ceil(each_bin))
```

```
for i in range(50):
    no_missing.iloc[i*math.ceil(each_bin) : i*math.ceil(each_bin)+math.ceil(each_bin),:] #bin
    this_mean = no_missing.iloc[i*math.ceil(each_bin) : i*math.ceil(each_bin)+math.ceil(each_bin),:]['avg_price'].mean() #  mean  bin
    no_missing.iloc[i*math.ceil(each_bin) : i*math.ceil(each_bin)+math.ceil(each_bin),:] ['avg_price']=this_mean # mean  bin
no_missing
```
