# Looping (concat)

## 1.ดึงจังหวัดที่มีอยู่ทั้งหมดในตารางออกมา

```
all_city = set(new_table['city_name'])
print(len(all_city))
```

## 2.คำนวณค่า mean ของ collumn 'city_name' เพื่อเก็บไว้เติมใส่เมืองที่เป็น Missing ทุกร้าน

```
all_mean = new_table.dropna(subset = ['avg_price'])['avg_price'].mean()
all_mean
```

## 3.สร้าง สร้าง list ของตารางของทุกๆจังหวัด

```
all_cities_list = []
for each_city in sorted(list(all_city)):
  temp_table = new_table[new_table['city_name'] == each_city]
  temp_table[['avg_price']] = temp_table[['avg_price']].fillna(new_table[new_table['city_name'] == each_city].dropna()['avg_price'].mean())
  all_cities_list.append(temp_table)  
```

ตัวโชว์

```
pd.concat(all_cities_list).fillna(all_mean)
```

ทำให้ตัวโชว์เป็นตาราง

```
no_missing = pd.concat(all_cities_list).fillna(all_mean)
no_missing.head()
```

เลือกค่าเฉพาะส่วนของตาราง


```
Index_list = no_missing['avg_price'] > 1000
no_missing[Index_list]
```

# Smooth data (binning)

## HW10 ทำ Binning smooth by mean ค่า 'avg_price' ในตาราง no_missing แบ่ง 50 กลุ่ม (bins = 50)

panda เรียงค่า 
ใช้ iloc แบ่ง
เจออีกทีวันที่ 30 กันยายน 

```
import math
each_bin = no_missing.shape[0]/50
print(math.ceil(each_bin))
```

```
for i in range(50):
    no_missing.iloc[i*math.ceil(each_bin) : i*math.ceil(each_bin)+math.ceil(each_bin),:] #แบ่งbin
    this_mean = no_missing.iloc[i*math.ceil(each_bin) : i*math.ceil(each_bin)+math.ceil(each_bin),:]['avg_price'].mean() #หา  mean ของแต่ละ bin
    no_missing.iloc[i*math.ceil(each_bin) : i*math.ceil(each_bin)+math.ceil(each_bin),:] ['avg_price']=this_mean #เติมค่า mean ลงไปในแต่ละ bin
no_missing
```
